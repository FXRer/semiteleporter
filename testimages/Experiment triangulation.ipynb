## Calcul vectoriel d'une position sur l'image (2D)

* Le rayon lumineux passe par le centre de l'image à une distance `k` de l'objectif (position `(0,k)`)
* La source du rayon lumineux est en position `(i, j)`
* Le point observé est `p`
* L'angle entre le centre de l'image et le point observé est `θ`

### Principe

    a = (i,j) (position du laser)
    b = (0,k) (point où le laser est au centre de l'image)
      son perpendiculaire B = (k,0)
    Rayon du laser: f(λ) = a + λ(b-a) = (i,j) + λ(-i,k-j)
    Rayon "sortant" de la caméra passant par ce point: f(μ) = μ(b + tan(θ).B) = μ(k.tan(θ),k)

Nous savons que le point est à l'intersection du laser et du rayon sortant de la caméra, donc

$
\begin{pmatrix} i \\ j \end{pmatrix} + 
\lambda \begin{pmatrix} -i \\ k-j \end{pmatrix} - 
\mu \begin{pmatrix} k\tan(\theta) \\ k \end{pmatrix} 
\Leftrightarrow
\begin{pmatrix} i \\ j \end{pmatrix} - 
\lambda \begin{pmatrix} i \\ j-k \end{pmatrix} - 
\mu \begin{pmatrix} k\tan(\theta) \\ k \end{pmatrix} 
\Leftrightarrow 
\begin{pmatrix} i & k \tan(\theta) \\ j-k & k \end{pmatrix}
\begin{pmatrix} \lambda \\ \mu \end{pmatrix} = 
\begin{pmatrix} i \\ j \end{pmatrix}
$

```python
from math import tan
k = 58.0            # Distance de la camera au point ou le laser est au centre de l'image (cm)
i, j = -45.0, 26.5  #Position du laser par rapport à la camera

A, B = np.array([i,j]), np.array([0,k])
Ax, Ay = A
Bx, By = B

# (x,y)
def pos(theta):
    matrix = np.array([[-i, k*tan(theta)], [k-j, k]])
    l, m = np.linalg.solve(matrix, A)
    return A + l*(B-A)
    #return m * np.array([k*tan(theta), k])
```

## Calcul de l'angle (θ)

* Notons `φ` l'angle maximal entre le centre et le bord (droit) de l'image
* Notons `E` la largeur de l'image en pixels, `e` la distance en pixels entre le centre de l'image et l'image de `p` 
* Notons `m` la largeur réelle entre les 2 bords observables en `b`

### Principe

    m = 2*k*tan(φ)  => φ = atan(m/2k)
    θ = φ * e/(E/2)

```python
from math import atan, pi
m = 69
phi = atan(float(m)/(2*k))
E = diff.shape[1]
print "Ouverture camera:", 180*phi/pi, "deg"

theta = lambda e: phi * e/(E/2)

X, Y = [], []
XY = []
i = 260
j_save = 0
for j in range(diff.shape[1]):
    if tuple(diff[i][j]) != (0, 0, 0):
        j_save = j
        t = theta(j - E/2)
        XY.append(pos(t))
        print 180*t/pi, XY[-1]

X, Y = map(np.array, zip(*XY))
```

### Visualisation du résultat

Affichage sur un plan

* Rayon laser (équation vectorielle établie à partir des mesures), en rouge
* Rayon capté par la caméra, en jaune
* Champ de vue de la caméra (vert)
* Point détecté (points bleus)
* Multiprise (en noir)

```python
fig = plt.subplot(111, aspect='equal')
# Computed points from above
fig.scatter(X, Y)

#shortcut for vector
vec = lambda *args : np.array(args)

def line(direction, position=np.array([0, 0]), **kwargs):
    """Draw a parametric line"""
    xy = [position + t*direction for t in np.linspace(0, 1)]
    x, y = zip(*xy)
    fig.plot(x, y, **kwargs)

# Image background
line(vec(m, 0), position=vec(-m/2, k), color='black', linewidth=3)
    
# A ray from the camera
line(vec(tan(theta(j_save - E/2))*k, k), color='y')

# Camera left limit
line(vec(tan(theta(-E/2))*k, k), color='g')

# Camera right limit
line(vec(tan(theta(E/2))*k, k), color='g')

#Laser ray
line(B-A, A, color='r')

#System parameters
line(B, color='m', ls='--')
line(A*vec(1, 0), color='m', ls='--')
line(A*vec(0, 1), A*vec(1, 0), color='m', ls='--')

fig.set_aspect('equal')
plt.show()
plt.imshow(img_red)
```
